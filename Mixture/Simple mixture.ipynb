```python
import numpy as np
import pymc3 as pm

with pm.Model() as model:
    dist = pm.Mixture('dist', np.array([0.25, 0.75]),
                      [pm.Uniform.dist(-0.15, -0.05), pm.Uniform.dist(0., 0.05)])

N = 40000
samples = dist.random(size=N)
```

```python
plt.hist(samples, 100, density=True);
```
